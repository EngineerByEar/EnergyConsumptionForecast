```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weather = pd.read_csv(
    "./weather_data/weather_data.csv",
    parse_dates=["DateUTC"],
    index_col="DateUTC"
)

energy = pd.read_csv(
    "./energy_data/energy_data.csv",
    parse_dates=["DateUTC"],
    index_col="DateUTC"
)
```

```python
weather.index = weather.index.tz_convert(None)
```

```python
data = pd.merge(
    energy,
    weather,
    left_index=True,
    right_index=True,
    how="left"
)
```

```python
data = pd.merge(
    pd.get_dummies(data["weekday"]).astype(int), 
    data,
    left_index = True,
    right_index = True,
    how = "left"
)
```

```python
data = data.drop(["weekday"], axis = 1)
```

```python
data.to_csv("../energy_and_weather_data.csv")
```

```python
num_cols = ['Value', 'hour', 'month', 'lag_1', 'lag_24',
       'lag_168', 'rolling_mean_24', 'rolling_std_24', 'rolling_mean_168',
       'rolling_std_168', 'is_free_day', 'ff', 'rr', 'so_h', 'tl', 'temp_24h',
       'temp_72h', 'sun_24h', 'hdg', 'cdg']
```

```python
corr = data[num_cols].corr()
```

```python
plt.figure(figsize=(12, 10))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    square=True,
    linewidths=1
)

plt.title("Feature Correlation Heatmap")
plt.savefig("Feature Corellation Heatmap")
plt.show()
```
